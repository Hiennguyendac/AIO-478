```python
! gdown 1iA0WmVfW88HyJvTBSQDI5vesf-pgKabq
import pandas as pd

# Tải dữ liệu
df = pd.read_csv('/content/advertising.csv')

# Chuyển DataFrame thành mảng NumPy
data = df.to_numpy()

# Câu hỏi 15: Giá trị lớn nhất và chỉ mục của nó trong cột Sales
max_sales = df['Sales'].max()
index_max_sales = df['Sales'].idxmax()

print("Max:", max_sales, "- Index:", index_max_sales)

# Câu hỏi 16: Giá trị trung bình của cột TV
mean_tv = df['TV'].mean()

print("Giá trị trung bình của cột TV là:", mean_tv)
```

```python
# Đếm số lượng bản ghi có giá trị tại cột Sales lớn hơn hoặc bằng 20
num_sales_ge_20 = (df['Sales'] >= 20).sum()
num_sales_ge_20
```

```python
# Lọc các bản ghi có giá trị tại cột Sales lớn hơn hoặc bằng 15
filtered_df = df[df['Sales'] >= 15]

# Tính giá trị trung bình của cột Radio cho các bản ghi đã lọc
mean_radio_filtered = filtered_df['Radio'].mean()
mean_radio_filtered
```

```python
# Tính giá trị trung bình của cột Newspaper
mean_newspaper = df['Newspaper'].mean()

# Lọc các bản ghi có giá trị tại cột Newspaper lớn hơn giá trị trung bình
filtered_df_newspaper = df[df['Newspaper'] > mean_newspaper]

# Tính tổng của cột Sales cho các bản ghi đã lọc
total_sales_filtered = filtered_df_newspaper['Sales'].sum()
total_sales_filtered
```

```python
# Tính giá trị trung bình của cột Sales
mean_sales = df['Sales'].mean()

# Tạo mảng mới scores dựa trên điều kiện cho trước
scores = ['Good' if x > mean_sales else 'Bad' if x < mean_sales else 'Average' for x in df['Sales']]

# In kết quả scores từ index 7 đến 10
scores[7:10]
```

```python
# Tính giá trị trung bình của cột Sales
mean_sales = df['Sales'].mean()

# Tìm giá trị trên cột Sales gần nhất với giá trị trung bình
closest_to_mean = df['Sales'].iloc[(df['Sales'] - mean_sales).abs().argsort()[:1]].values[0]

# Tạo mảng mới scores dựa trên điều kiện cho trước
scores = ['Good' if x > closest_to_mean else 'Bad' if x < closest_to_mean else 'Average' for x in df['Sales']]

# In toàn bộ mảng scores để kiểm tra
print(scores[7:10])
```

```python
# Download image
!gdown 1i9dqan21DjQoG5Q_VEvm0LrVwAlXD0vB

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

# Đọc hình ảnh
img = mpimg.imread('/content/dog.jpeg')

# Chuyển đổi hình ảnh thành ảnh xám bằng phương pháp lightness
gray_img_01 = (np.max(img, axis=2) + np.min(img, axis=2)) / 2

# Hiển thị hình ảnh gốc và ảnh xám
plt.figure(figsize=(10, 5))

# Hiển thị hình ảnh gốc
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

# Hiển thị hình ảnh xám
plt.subplot(1, 2, 2)
plt.imshow(gray_img_01, cmap='gray')
plt.title('Grayscale Image (Lightness Method)')
plt.axis('off')

# Hiển thị cả hai hình ảnh
plt.show()

# In giá trị pixel đầu tiên của ảnh xám
print("Giá trị pixel đầu tiên của ảnh xám:", gray_img_01[0, 0])
```
